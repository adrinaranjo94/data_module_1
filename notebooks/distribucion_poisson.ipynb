```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Par치metro
lambda_param = 3  # tasa media de ocurrencia

# Generar datos con distribuci칩n Poisson
data = poisson.rvs(lambda_param, size=1000)
df = pd.DataFrame(data, columns=['Values'])

# Visualizaci칩n
sns.histplot(df['Values'], discrete=True, kde=False)
plt.title('Distribuci칩n Poisson')
plt.show()
```
