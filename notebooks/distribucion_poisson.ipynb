```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Parámetro
lambda_param = 3  # tasa media de ocurrencia

# Generar datos con distribución Poisson
data = poisson.rvs(lambda_param, size=1000)
df = pd.DataFrame(data, columns=['Values'])

# Visualización
sns.histplot(df['Values'], discrete=True, kde=False)
plt.title('Distribución Poisson')
plt.show()
```
