```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import lognorm

# Parámetros
sigma = 1  # desviación estándar logarítmica

# Generar datos con distribución log-normal
data = lognorm.rvs(sigma, size=1000)
df = pd.DataFrame(data, columns=['Values'])

# Visualización
sns.histplot(df['Values'], kde=True)
plt.title('Distribución Log-normal')
plt.show()
```
